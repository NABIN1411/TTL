```python
#1.From the given dataset print the first and last five rows
import pandas as pd
df=pd.read_csv('Automobile_data.csv')
df.head(5)
```

```python
df.tail(5)
```

```python
#2.Print All Volkswagen Cars details and also sort them from highest to lowest price.
volkswagen_cars = df[df['company'] == 'volkswagen'].sort_values(by='price', ascending=False)
print(volkswagen_cars)
```

```python
#3.Count total cars per company and their average price
total_cars = df.groupby('company')['company'].count()
avg_price = df.groupby('company')['price'].mean()
print(total_cars)
print(avg_price)
```

```python
#4.Find the average mileage of each car making company
avg_mileage = df.groupby('company')['average-mileage'].mean()
print(avg_mileage)
```

```python
#5.Create two data frames, Merge two data frames, and append the second data frame as a new column to the first data frame.
df1 = df[['index', 'company', 'body-style']]
df2 = df[['index', 'wheel-base', 'length']]
merged_df = pd.merge(df1, df2, on='index')
print(merged_df)
```

```python
appended_df = df1.join(df2.set_index('index'), on='index')
print(appended_df)
```
